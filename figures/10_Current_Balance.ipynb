```julia
using Plots
using LaTeXStrings
pgfplots()
```

```julia
cur_x = 1.25*[-1,1,1,-1,-1]
cur_y = 1*[1,1,-1,-1,1]

cur_plot = plot()

plot!(cur_x, cur_y+2.5, color=1, fill=true, fillopacity=0.05, label="")
plot!(cur_x-3.25, cur_y-2.5, color=1, fill=true, fillopacity=0.05, label="")
plot!(cur_x+3.25, cur_y-2.5, color=1, fill=true, fillopacity=0.05, label="")
ylims!(-6,6)
plot!(aspectratio=:equal)

p = pi/3 - pi/32

r = 6
a = 2.75

tt = linspace(0, 2*pi, 200)

x(t) = ( r * cos(p) * cos(t) - a * sin(p) * sin(t) )
y(t) = ( r * sin(p) * cos(t) + a * cos(p) * sin(t) ) 

tmp_x = x.(tt)
tmp_y = y.(tt)

plot!(tmp_x - 3.25/2, tmp_y, color=2, label="Steady State")
plot!(-tmp_x + 3.25/2, tmp_y, color=3, label="Pure Pulsed")

plot!(ann=[(0,2.5,L"f_{BS}")])
plot!(ann=[(-3.25,-2.5,L"f_{CD}")])
plot!(ann=[(+3.25,-2.5,L"f_{ID}")])

plot!(grid=false, ticks=false)
title!("Current Balance")
```

```julia
savefig(cur_plot, "../images/current_balance.tex")
```
