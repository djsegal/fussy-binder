```julia
cur_indices

plot_one = plot(aspectratio=:equal)

cur_r = 5

bad_indices = [ 13 , 7 , 19 , 3 , 16 , 23 ]
good_indices = [ 10 ]

for cur_index in 0:cur_r
    plot!([1,1]*cur_index,[0,cur_r],color=1,label="")
    plot!([0,cur_r],[1,1]*cur_index,color=1,label="")
end

for (cur_super_index, cur_indices) in enumerate([bad_indices,good_indices])
    for (cur_position, cur_index) in enumerate(cur_indices)
        cur_i = 1 + Int(floor((cur_index-1)/cur_r))
        cur_j = 1 + mod((cur_index-1),cur_r)

        cur_i *= -1
        cur_i += cur_r + 1

        plot!(
            cur_j + [0, 0, 1, 1, 0] - 1,
            cur_i + [0, 1, 1, 0, 0] - 1,
            fill=true, color=cur_super_index+1, linewidth=0, fillopacity=0.7, 
            label= cur_position == 1 ? label= cur_super_index == 1 ? "Invalid Reactor" : "Valid Reactor" : ""
        )
    end
end

plot!(xticks=([0.5,1.5,2.5,3.5,4.5,5.5],[1,2,3,4,5]),yticks=([0.5,1.5,2.5,3.5,4.5,5.5],[1,2,3,4,5]),grid=false)

ylims!(-2,cur_r+2)

ylabel!("Row")
xlabel!("Column")

title!("Step I - Find a Valid Reactor")

plot!()
```

```julia
plot_two = plot()
max_iteration = 5

y_bounds = [NaN,NaN]

for cur_iteration in 1:max_iteration
    cur_y = 2 ^ max_iteration / 2 ^ (cur_iteration-1)
    if cur_iteration == 1
        cur_label = "Valid Reactor"
        cur_color = 3
        y_bounds[1] = cur_y
    elseif cur_iteration == max_iteration
        cur_label = "Invalid Reactor"
        cur_color = 2
        y_bounds[2] = cur_y
    else
        cur_label = ""
        cur_color = 3
        y_bounds[1] = cur_y
    end
    scatter!([cur_iteration],[cur_y],markersize=8,label=cur_label,color=cur_color)
end
    
xlims!(0,6)

xlabel!("Iteration")
ylabel!("Cost")

plot!(xticks=[1,2,3,4,5], yticks=([12,24,36],[1,2,3]))
ylims!(0,40)

plot!([-1,-1,7,7,-1],[y_bounds...,reverse(y_bounds)...,y_bounds[1]],fill=true,color=1,opacity=0.15,label="Bounded Cost")

title!("Step II - Bound Minimum Cost Reactor")

plot!()
```

```julia
plot_three = plot()

xlims!(0,6)
plot!(xticks=[1,2,3,4,5],yticks=(1:5)./10)

cur_width = 0.15

cur_min = 0.1
cur_max = 0.5

cur_index = 1
plot!(
    [cur_index-cur_width, cur_index-cur_width,cur_index+cur_width,cur_index+cur_width,cur_index-cur_width],
    [cur_min,cur_max,cur_max,cur_min,cur_min],
    fill=true, opacity=0.3, color=1, label=""
)

plot!(
    [cur_index-cur_width,cur_index+cur_width],
    [cur_max,cur_max],
    color=3, label="Valid Reactor", width=2
)

plot!(
    [cur_index-cur_width,cur_index+cur_width],
    [cur_min,cur_min],
    color=2, label="Invalid Reactor", width=2
)

for (cur_super_index,cur_switch) in enumerate([1,0,0,1])

    if iszero(cur_switch)
        cur_min = middle(cur_min, cur_max)
    else
        cur_max = middle(cur_min, cur_max)
    end

    cur_index += 1

    plot!(
        [cur_index-cur_width, cur_index-cur_width,cur_index+cur_width,cur_index+cur_width,cur_index-cur_width],
        [cur_min,cur_max,cur_max,cur_min,cur_min],
        fill=true, opacity=0.3, color=1, label=cur_super_index == 1 ? "Bounded Cost" : ""
    )

    plot!(
        [cur_index-cur_width,cur_index+cur_width],
        [cur_min,cur_min],
        color=2, label="", width=2
    )

    plot!(
        [cur_index-cur_width,cur_index+cur_width],
        [cur_max,cur_max],
        color=3, label="", width=2
    )
end
    
ylims!(0,0.6)

ylabel!("Cost")
xlabel!("Iteration")

title!("Step III - Hone Minimum Cost Reactor")

plot!()
```
