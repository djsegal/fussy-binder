```julia
funding_dict = OrderedDict()

cur_years = [ 1976,1977,1978,1979,1980,1981,1982,1983,1984,1985,1986,1987,1988,1989,1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005 ]

funding_dict["Moderate"] = [159,274,303,291,329,386,403,389,413,415,506,573,572,567,613,719,682,663,689,600,571,582,683,824,855,836,807,728,639,600]
funding_dict["Aggressive"] = [159,304,410,430,427,648,755,715,648,686,794,835,879,875,746,657,728,899,1030,1041,852,703,654]
funding_dict["Accelerated"] = [159,329,437,520,709,911,949,836,1013,1106,1132,1016,937,1018,1329,1180,931,742]
funding_dict["Maximum"] = [159,379,650,1285,1627,1605,2016,2196,1829,1146,1303,1890,2151,1446,985]

dollar_converter = 3.86 / 1000

cur_plot = plot()

other_years = [1985,1986,1987,1988,1989,1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
other_funding = [1001836,827367,754179,702929,697497,608102,527899,593976,567297,545463,545963,382175,351267,343818,325482,347117,342814,336267,328250,339832,342925,349302,377075,343662,461341,480510,409602,429380,406833,534818,494834,457833,389125,377419]

other_funding /= 1e6

plot!(other_years, other_funding, label="Actual Funding", color=1)

plot!([1977, 2018],[1,1], style=:dot, label="``Fusion Never''", color=:black, opacity=0.7)

for (cur_index, (cur_key, cur_value)) in enumerate(funding_dict)
    tmp_y = cur_value
    tmp_x = cur_years[1:length(cur_value)]
    tmp_y *= dollar_converter
    plot!(tmp_x, tmp_y, color=cur_index+1, label="$cur_key Effort")
    scatter!(tmp_x[end:end], tmp_y[end:end], label="", color=cur_index+1)
end
plot!()

xlabel!("Year")
ylabel!("Annual Total Budget (billions 2018 \\\$)")
title!("Fusion Funding (Actual and 1976 Projections)")

ylims!(0,10)
xlims!(1975,2020)
```
